By reporting both micro-averaged F1 and macro-averaged Tag Recall, researchers present a more comprehensive assessment of their model's capabilities, acknowledging both the importance of overall performance (as assessed by micro F1) and the ability to handle all tags uniformly well (as assessed by Tag Recall).

Micro-Averaged F1: The micro-averaged F1 score aggregates the contributions of all classes to compute the average metric, which takes into account the frequency of each class. This metric is particularly useful when class imbalance is present, as it gives a weight to each instance of each class, thereby emphasizing the performance on the most common classes. It's a way to understand overall performance, especially when the correct prediction of each individual tag is important.
Tag Recall (Macro Recall): Tag Recall, as defined in the paper, seems to be analogous to what is commonly known as macro-averaged recall because it calculates the average recall per tag (class) without considering class imbalance. This means it treats each tag equally, no matter how many times it appears in the dataset.
